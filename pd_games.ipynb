```python
# 1. Обработка столбца year_of_release
# Заполняем пропуски медианным значением
median_year = df_news['year_of_release'].median()
df_news['year_of_release'] = df_news['year_of_release'].fillna(median_year)
```

```python
# Преобразуем в int64
df_news['year_of_release'] = df_news['year_of_release'].astype('int64')
```

```python
# 2. Обработка столбцов Sales
sales_columns = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
for col in sales_columns:
    # Заменяем нечисловые значения на NaN и преобразуем в float
    df_news[col] = pd.to_numeric(df_news[col], errors='coerce')
    # Заполняем пропуски нулями (отсутствие продаж)
    df_news[col] = df_news[col].fillna(0.0)
```

```python
# 3. Обработка critic_score
# Преобразуем в числовой тип, заменяя нечисловые значения на NaN
df_news['critic_score'] = pd.to_numeric(df_news['critic_score'], errors='coerce')
# Заполняем пропуски средним значением
mean_critic = df_news['critic_score'].mean()
df_news['critic_score'] = df_news['critic_score'].fillna(mean_critic)
```

```python
# 4. Обработка user_score
# Заменяем 'tbd' на NaN
df_news['user_score'] = df_news['user_score'].replace('tbd', 'pd.NaN')
# Преобразуем в числовой тип
df_news['user_score'] = pd.to_numeric(df_news['user_score'], errors='coerce')
# Заполняем пропуски средним значением
mean_user = df_news['user_score'].mean()
df_news['user_score'] = df_news['user_score'].fillna(mean_user)
```

```python
# 5. Проверка текстовых столбцов
text_columns = ['name', 'platform', 'genre', 'rating']
for col in text_columns:
    df_news[col] = df_news[col].astype('object')  # Приводим к типу object (текст)
```

```python
def convert_to_numeric(column, string_values_to_replace=None):
    if string_values_to_replace is None:
        string_values_to_replace = ['unknown', 'Unknown', 'UNKNOWN',
                           'n/a', 'N/A', 'na', '-', '.', '']
    column_cleaned = column.replace(string_values_to_replace)
    column_numeric = pd.to_numeric(column_cleaned, errors='coerce')
    return column_numeric
```

```python
# Список числовых столбцов
numeric_columns = [
    'year_of_release',
    'na_sales', 'eu_sales', 'jp_sales', 'other_sales',
    'critic_score', 'user_score'
]
```

## 4. Категоризация данных
    
Проведите категоризацию данных:
- Разделите все игры по оценкам пользователей и выделите такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
# Убедимся, что User Score имеет числовой тип
df_actual['User Score'] = pd.to_numeric(df_actual['User Score'], errors='coerce')

# Создаём функцию для категоризации оценок
def categorize_user_score(score):
    if pd.isna(score):  # Обработка пропусков
        return 'Unknown'
    elif 8 <= score <= 10:
        return 'Высокая оценка'
    elif 3 <= score < 8:
        return 'Средняя оценка'
    elif 0 <= score < 3:
        return 'Низкая оценка'
    else:
        return 'Некорректное значение'  # аномалий
```

```python
# Применяем функцию к столбцу User Score
df_actual['User Score Category'] = df_actual['User Score'].apply(categorize_user_score)
```

```python
print("=== РАСПРЕДЕЛЕНИЕ ИГР ПО ОЦЕНКАМ ПОЛЬЗОВАТЕЛЕЙ ===")
category_counts = df_actual['User Score Category'].value_counts()
category_percentages = df_actual['User Score Category'].value_counts(normalize=True) * 100

for category in ['Высокая оценка', 'Средняя оценка', 'Низкая оценка']:
    count = category_counts.get(category, 0)
    percent = category_percentages.get(category, 0.0)
    print(f"{category}: {count} игр ({percent:.2f}%)")

# Общая статистика
total_games = len(df_actual)
total_with_scores = total_games - category_counts.get('Unknown', 0)
print(f"\nВсего игр: {total_games}")
print(f"Из них с оценками пользователей: {total_with_scores} ({(total_with_scores / total_games * 100):.2f}%)")
```

- Разделите все игры по оценкам критиков и выделите такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
# Создаём функцию для категоризации оценок критиков
def categorize_critic_score(score):
    if pd.isna(score):  # Обработка пропусков
        return 'Unknown'
    elif 80 <= score <= 100:
        return 'Высокая оценка'
    elif 30 <= score < 80:
        return 'Средняя оценка'
    elif 0 <= score < 30:
        return 'Низкая оценка'
    else:
        return 'Некорректное значение'  # анамалии 

```

```python
# Применяем функцию к столбцу Critic Score
df_actual['Critic Score Category'] = df_actual['Critic Score'].apply(categorize_critic_score)
```

```python
print("=== РАСПРЕДЕЛЕНИЕ ИГР ПО ОЦЕНКАМ КРИТИКОВ ===")
category_counts = df_actual['Critic Score Category'].value_counts()
category_percentages = df_actual['Critic Score Category'].value_counts(normalize=True) * 100


for category in ['Высокая оценка', 'Средняя оценка', 'Низкая оценка', ]:
    count = category_counts.get(category, 0)
    percent = category_percentages.get(category, 0.0)
    print(f"{category}: {count} игр ({percent:.2f}%)")
```
